```python
import pandas as pd
import os
from PIL import Image
from tqdm import tqdm
```

```python
train_csv_path = "../data/train2.csv"
train_path = "../data/train/"
meta_data_path = "../data/meta.csv"

train_df = pd.read_csv(train_csv_path)
meta_data = pd.read_csv(meta_data_path)
train_df
```

```python
widths, heights = [], []

for img_id in tqdm(train_df["ID"], desc="이미지 크기 계산"):
    img_path = os.path.join(train_path, img_id)
    with Image.open(img_path) as img:
        w, h = img.size
    widths.append(w)
    heights.append(h)

train_df["width"] = widths
train_df["height"] = heights

# aspect_ratio, aspect_bin, strata 컬럼 생성
train_df["aspect_ratio"] = train_df["width"] / train_df["height"]
train_df["aspect_bin"] = pd.cut(train_df["aspect_ratio"], bins=4, labels=False)
train_df["strata"] = train_df["target"].astype(str) + "_" + train_df["aspect_bin"].astype(str)
train_df
```
